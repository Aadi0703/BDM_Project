```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/content/data2 (1).csv')
```

```python
data.head(40)
```

```python
# Drop rows where all the columns are empty
cleaned_data = data.dropna(how='all')
```

```python
cleaned_data.head(30)
```

```python
# Forward fill (ffill) the first 3 columns where they are empty
columns_to_ffill = ['Bill No.', 'Date', 'Party Name']
cleaned_data[columns_to_ffill] = cleaned_data[columns_to_ffill].ffill()
```

```python
cleaned_data.head(30)
```

```python
# Create the first new column based on the condition for "Party Name"
cleaned_data['Party Type'] = cleaned_data['Party Name'].apply(
    lambda x: 'Government' if x == 'THE DEAN, SIR JJ GROUP OF HOSPITAL' else 'Private'
)

# Create the second new column for the month extracted from the "Date" column
cleaned_data['Month'] = pd.to_datetime(cleaned_data['Date'], format='%m-%d-%Y', errors='coerce').dt.month
```

```python
cleaned_data.head(30)
```

```python
# Replaced all occurrences of '40-mm-yyyy' with '14-mm-yyyy' in the 'Date' column
cleaned_data['Date'] = cleaned_data['Date'].str.replace(r'40-(\d{2}-\d{4})', r'14-\1', regex=True)

# Recalculated the "Month" column based on the updated "Date" column
cleaned_data['Month'] = pd.to_datetime(cleaned_data['Date'], format='%d-%m-%Y', errors='coerce').dt.month_name()
```

```python
cleaned_data = cleaned_data.iloc[233:]
```

```python
cleaned_data.head(10)
```

```python
cleaned_data = cleaned_data.drop(['Bill No.', 'Batch No'], axis=1)
cleaned_data.head(10)
```

```python
import matplotlib.pyplot as plt

# Convert "Amount" to numeric after removing commas, handling errors gracefully
cleaned_data['Amount'] = pd.to_numeric(cleaned_data['Amount'].str.replace(',', ''), errors='coerce')

# Group data by "Month" and calculate the sum of "Amount"
revenue_by_month = cleaned_data.groupby('Month')['Amount'].sum()

# Filter the months from February to September and sort them in the correct order
month_order = ['February', 'March', 'April', 'May', 'June', 'July', 'August', 'September']
filtered_revenue = revenue_by_month.loc[month_order]

# Plotting the bar graph
plt.figure(figsize=(10, 6))
filtered_revenue.plot(kind='bar')
plt.title('Revenue Generated from February to September')
plt.xlabel('Month')
plt.ylabel('Total Revenue (Amount)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting the bar graph with improved aesthetics and custom Y-axis labels
plt.figure(figsize=(10, 6))
filtered_revenue.plot(kind='bar', color='skyblue', edgecolor='black')

# Adding title and labels with improved styling
plt.title('Revenue Generated from February to September', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Revenue (Amount)', fontsize=14)

# Customizing Y-axis labels to show in the format 10000, 12000, etc.
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{int(x):,}'))

# Rotating X-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# prompt: give me a line chart of above graph start the y axis with 0

# Plotting the line chart with improved aesthetics and custom Y-axis labels
plt.figure(figsize=(10, 6))
filtered_revenue.plot(kind='line', marker='o', color='skyblue', linewidth=2, markersize=8)

# Adding title and labels with improved styling
plt.title('Revenue Generated from February to September', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Revenue (Amount)', fontsize=14)

# Customizing Y-axis labels to show in the format 10000, 12000, etc.
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{int(x):,}'))
plt.ylim(ymin=0) # Start y-axis from 0

# Rotating X-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adding a grid for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Grouping the data by Month and Party Type, summing up the revenue for each combination
monthly_revenue_by_type = cleaned_data.groupby(['Month', 'Party Type'])['Amount'].sum().unstack()

# Filtering and sorting the months in the correct order
filtered_monthly_revenue = monthly_revenue_by_type.loc[month_order]

# Plotting the stacked bar chart for Government and Private
filtered_monthly_revenue.plot(kind='bar', stacked=True, figsize=(10, 6), color=['green', 'red'], edgecolor='black')

# Adding title and labels
plt.title('Monthly Revenue Distribution by Category', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Revenue (Amount)', fontsize=14)

# Customizing Y-axis labels
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{int(x):,}'))

# Adding legend
plt.legend(title="Category", fontsize=12)

# Rotating X-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Summing up the total revenue for Government and Private categories
category_revenue = cleaned_data.groupby('Party Type')['Amount'].sum()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
category_revenue.plot(
    kind='pie',
    autopct='%1.2f%%',
    startangle=90,
    colors=['green', 'red'],
    labels=['Government', 'Private'],
    explode=(0.1, 0),  # To highlight the Government category
)

# Adding title
plt.title('Revenue Distribution by Category', fontsize=16, fontweight='bold')

# Ensuring the plot is circular
plt.axis('equal')

# Display the pie chart
plt.tight_layout()
plt.show()
```

```python
# Filter the dataset for relevant months and products
products_of_interest = [
    "MECROSS NC BALLOON",
    "HEART STENTS",
    "NASAL PRONGS",
    "Heat and Moisture Exchange/HME filter",
    "MICRO INFUSION SET",
    "HALE CISER",
    "MASKS",
    "3 way Stop Cock"
]

# Filter data for products of interest and months from February to September
filtered_data = cleaned_data[cleaned_data['Item Name'].isin(products_of_interest) &
                             cleaned_data['Month'].isin(['February', 'March', 'April', 'May', 'June', 'July', 'August', 'September'])]

# Group data by Month and Item Name, summing up quantities
sales_data = filtered_data.groupby(['Month', 'Item Name'])['Qty.'].sum().unstack(fill_value=0)

# Sort the data by the order of months
sales_data = sales_data.loc[month_order]

# Plot the line graph with different colors for each product
plt.figure(figsize=(12, 8))
sales_data.plot(kind='line', marker='o', figsize=(12, 8))

plt.title('Monthly Sales of Selected Products (February to September)', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Quantity Sold', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title="Product", fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Group by month and product, summing the quantities
monthly_product_sales = filtered_data.groupby(['Month', 'Item Name'])['Qty.'].sum().unstack(fill_value=0)

# Reorder the months to follow February to September order
monthly_product_sales = monthly_product_sales.loc[
    ['February', 'March', 'April', 'May', 'June', 'July', 'August', 'September']
]
# Print the DataFrame for reference
print(monthly_product_sales)
```

```python
# Filter data for the specified products of interest
products_of_interest = [
    "MECROSS NC BALLOON",
    "HEART STENTS",
    "NASAL PRONGS",
    "Heat and Moisture Exchange/HME filter",
    "MICRO INFUSION SET",
    "HALE CISER",
    "MASKS",
    "3 way Stop Cock"
]

# Filter the data for relevant products
filtered_revenue_data = cleaned_data[cleaned_data['Item Name'].isin(products_of_interest)]

# Calculate total revenue for each product
product_revenue = filtered_revenue_data.groupby('Item Name')['Amount'].sum()

# Plot the pie chart showing percentage revenue distribution
plt.figure(figsize=(10, 8))
product_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Percentage Revenue from Products of Interest', fontsize=16, fontweight='bold')
plt.ylabel('')  # Remove the default y-label
plt.tight_layout()
plt.show()
```

```python
# Create subplots for each product
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.flatten()  # Flatten the axes array for easier iteration

for i, product in enumerate(monthly_product_sales.columns):
    axes[i].bar(monthly_product_sales.index, monthly_product_sales[product])
    axes[i].set_title(f'Monthly Sales of {product}')
    axes[i].set_xlabel('Month')
    axes[i].set_ylabel('Number of Products')
    axes[i].tick_params(axis='x', rotation=45)

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()
plt.show()
```

```python
# Create subplots for each month (February to September)
months = ['February', 'March', 'April', 'May', 'June', 'July', 'August', 'September']
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.flatten()

for i, month in enumerate(months):
    # Filter data for the current month
    monthly_data = monthly_product_sales.loc[[month]]

    # Create the bar plot
    axes[i].bar(monthly_data.columns, monthly_data.iloc[0])
    axes[i].set_title(f'Product Sales in {month}')
    axes[i].set_xlabel('Products')
    axes[i].set_ylabel('Number of Products Sold')
    axes[i].tick_params(axis='x', rotation=45) # Rotate x-axis labels for better readability

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()
plt.show()
```
